```python
import pandas as pd
import numpy as np

bahrain='https://www.formula1.com/en/results.html/2021/races/1064/bahrain/race-result.html'
emilia_romagna='https://www.formula1.com/en/results.html/2021/races/1065/italy/race-result.html'
portugal='https://www.formula1.com/en/results.html/2021/races/1066/portugal/race-result.html'
spain='https://www.formula1.com/en/results.html/2021/races/1086/spain/race-result.html'
monaco='https://www.formula1.com/en/results.html/2021/races/1067/monaco/race-result.html'
azerbaijan='https://www.formula1.com/en/results.html/2021/races/1068/azerbaijan/race-result.html'
france='https://www.formula1.com/en/results.html/2021/races/1070/france/race-result.html'
styria='https://www.formula1.com/en/results.html/2021/races/1092/austria/race-result.html'
austria='https://www.formula1.com/en/results.html/2021/races/1071/austria/race-result.html'
great_britain='https://www.formula1.com/en/results.html/2021/races/1072/great-britain/race-result.html'
hungary='https://www.formula1.com/en/results.html/2021/races/1073/hungary/race-result.html'
belgium='https://www.formula1.com/en/results.html/2021/races/1074/belgium/race-result.html'
netherlands='https://www.formula1.com/en/results.html/2021/races/1075/netherlands/race-result.html'
italy='https://www.formula1.com/en/results.html/2021/races/1076/italy/race-result.html'
russia='https://www.formula1.com/en/results.html/2021/races/1077/russia/race-result.html'
turkey='https://www.formula1.com/en/results.html/2021/races/1078/turkey/race-result.html'
united_states='https://www.formula1.com/en/results.html/2021/races/1102/united-states/race-result.html'
mexico='https://www.formula1.com/en/results.html/2021/races/1103/mexico/race-result.html'
brazil='https://www.formula1.com/en/results.html/2021/races/1104/brazil/race-result.html'
qatar='https://www.formula1.com/en/results.html/2021/races/1105/qatar.html'
saudi_arabia='https://www.formula1.com/en/results.html/2021/races/1106/saudi-arabia/race-result.html'
abu_dhabi='https://www.formula1.com/en/results.html/2021/races/1107/abu-dhabi/race-result.html'
def read_html(link):
# Read the HTML file.
    file= pd.read_html(link)
# Select the first table.
    table= file[0]
    table=table.loc[:,['Pos','Driver','Car','PTS']] #Choose the columns that we are interested in.
    return table



print(read_html(bahrain))
countries=['bahrain','emilia_romagna','italy','portugal','spain','monaco','azerbaijan','france','styria','austria','great_britain','hungary','belgium','netherlands','italy','russia','turkey','united_states','mexico','brazil','saudi_arabia','abu_dhabi']

```

This function generates an adjacency matrix for Formula 1 drivers based on race results. It assigns points to drivers who finished ahead of each other, handles tiebreakers, and returns a matrix reflecting head-to-head performance

```python
def adj_matrix(data):
    dat=pd.DataFrame(np.zeros([20,20]),index=read_html(bahrain).loc[:,'Driver'],columns=read_html(bahrain).loc[:,'Driver'],dtype=int)
    ranking=[]
    for driver in np.array(data.loc[:,'Driver']):
        for drivers in read_html(bahrain).loc[:,'Driver']:
            if drivers not in ranking:
                dat.loc[driver,drivers]=1
                ranking.append(driver)
                dat.loc[driver,driver]=0 #0 points to themselves
            continue

    return dat
    for name in ranking:
        if data.loc[name,'PTS']==0:
            tie.append(name)
    for i in tie:
        for j in tie:
            dat.loc[i,j]==1
            dat.loc[i,i]==0

    return dat
print(adj_matrix(read_html(italy)))
```

This part accumulates and updates a driver adjacency matrix across multiple Formula 1 races, reflecting their head-to-head performance.

```python
updated=adj_matrix(read_html(bahrain))+adj_matrix(read_html(italy))
updated1=updated+adj_matrix(read_html(portugal))+adj_matrix(read_html(spain))
updated2=updated1+adj_matrix(read_html(monaco))+adj_matrix(read_html(azerbaijan))
updated3=updated2+adj_matrix(read_html(france))+adj_matrix(read_html(austria))
updated4=updated3+adj_matrix(read_html(great_britain))+adj_matrix(read_html(hungary))
updated5=updated4+adj_matrix(read_html(belgium))+adj_matrix(read_html(russia))
updated6=updated5+adj_matrix(read_html(turkey))+adj_matrix(read_html(united_states))
updated7=updated6+adj_matrix(read_html(mexico))+adj_matrix(read_html(brazil))
updated8=updated7+adj_matrix(read_html(saudi_arabia))
updated9=updated8+adj_matrix(read_html(abu_dhabi))+adj_matrix(read_html(styria))
updated10=updated9+adj_matrix(read_html(emilia_romagna))+adj_matrix(read_html(netherlands))
updated=updated10.drop('Robert  Kubica  KUB',axis=1).drop('Robert  Kubica  KUB',axis=0)
updated=updated.T
print(updated)
```


Now it normalizes the accumulated driver adjacency matrix, updated, to create a stochastic matrix (df_stochastic). Each row in the matrix is divided by the sum of its elements, ensuring that the rows' sums equal 1.

```python

df_stochastic = updated.div(updated.sum(axis=1), axis=0) #normalize to get stocahstic/rows sum equal to 1
print(df_stochastic)
```


These functions implement the PageRank algorithm. The google_matrix function combines the stochastic matrix (stochastic_matrix) with a teleportation matrix using a damping factor (alpha). The power_method function iteratively calculates the dominant eigenvector of the resulting Google matrix (G) until convergence, with a specified tolerance (e). The final scores, representing the PageRank values, are obtained and printed.

```python
def google_matrix(stochastic_matrix,alpha): #google matrix with alpha
    E=pd.DataFrame(np.full((20,20), 1/20),columns=df_stochastic.index,index=df_stochastic.index)
    G=alpha*stochastic_matrix+(1-alpha)*E
    return G
print(google_matrix(df_stochastic,0.85))

def power_method(G,e):
    z=np.full((1,20),1/20)
    convergence=[]
    for i in range(20):
        q=np.dot(z,G)
        z=q
        convergence.append(q)
        if i>0:
            conv=np.linalg.norm(abs(q-convergence[i-1]),ord=1)
        continue
        print(i,conv)
        if conv<=e:
            break
    return convergence
```

This code computes the PageRank scores for Formula 1 drivers based on the final stochastic matrix (df_stochastic). It uses the power_method function with the Google matrix generated using a damping factor of 0.85. The resulting scores are organized into a DataFrame (final_scores) and printed, providing a ranked list of drivers based on their performance across the considered races.

```python
scores=np.array((power_method(np.matrix(google_matrix(df_stochastic,0.85)),0.001)[19]))
final_table=pd.DataFrame(scores.T,index=df_stochastic.index,columns=['Scores'])
final_scores=final_table.sort_values('Scores',ascending=False)
print(final_scores)

```
